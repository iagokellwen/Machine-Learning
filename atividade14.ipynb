Análise de Resultados de LDA
o	Após aplicar LDA em um conjunto de dados, analise como as classes estão separadas e discuta o impacto das variáveis na discriminação das classes.


```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis as LDA
```

```python
# 1. Carregar o conjunto de dados Iris
iris = load_iris()
X = iris.data
y = iris.target
target_names = iris.target_names
```

```python
# 2. Aplicar LDA
lda = LDA(n_components=2)  # Reduzir para 2 componentes
X_lda = lda.fit_transform(X, y)
```

```python
# 3. Visualizar as classes
plt.figure(figsize=(8, 6))
colors = ['navy', 'turquoise', 'darkorange']
lw = 2

for color, i, target_name in zip(colors, [0, 1, 2], target_names):
    plt.scatter(X_lda[y == i, 0], X_lda[y == i, 1], color=color, alpha=.8, lw=lw, label=target_name)
plt.title('LDA dos Dados Iris')
plt.xlabel('Componente Discriminante 1')
plt.ylabel('Componente Discriminante 2')
plt.legend(loc='best', shadow=False, scatterpoints=1)
plt.grid()
plt.show()
```

```python
# 4. Analisar o impacto das variáveis
coefficients = lda.coef_
print("Coeficientes das variáveis nas componentes discriminantes:")
for i, target_name in enumerate(target_names):
    print(f"{target_name}: {coefficients[i]}")

# Análise da variância explicada
explained_variance = lda.explained_variance_ratio_
print("Variância explicada pelas componentes discriminantes:")
print(explained_variance)
```
