LDA com Scikit-Learn
o	Use o LDA da biblioteca Scikit-Learn em um conjunto de dados e avalie a precisão do modelo resultante em uma tarefa de classificação.


```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis as LDA
from sklearn.metrics import accuracy_score
```

```python
# 1. Carregar o conjunto de dados Iris
iris = load_iris()
data_iris = iris.data
labels = iris.target
```

```python
# 2. Dividir os dados em conjuntos de treinamento e teste
X_train, X_test, y_train, y_test = train_test_split(data_iris, labels, test_size=0.2, random_state=42)
```

```python
# 3. Aplicar o LDA
lda = LDA()
lda.fit(X_train, y_train)
```

```python
# 4. Fazer previsões usando os dados de teste
y_pred = lda.predict(X_test)
```

```python
# 5. Avaliar a precisão do modelo
accuracy = accuracy_score(y_test, y_pred)
```

```python
# 6. Exibir a precisão
print(f'Precisão do modelo LDA: {accuracy * 100:.2f}%')
```

```python
# 7. Visualizar os dados LDA
data_lda = lda.transform(X_train)
```

```python
plt.figure(figsize=(8, 6))
for i, label in enumerate(np.unique(y_train)):
    plt.scatter(data_lda[y_train == label, 0], np.zeros_like(data_lda[y_train == label, 0]),
                label=iris.target_names[label], alpha=0.7)

plt.title('Dados Iris Projetados pelo LDA (Treinamento)')
plt.xlabel('Discriminante Linear 1')
plt.ylabel('Discriminante Linear 2 (zero devido à 1D)')
plt.legend(loc='best')
plt.tight_layout()
plt.show()
```
