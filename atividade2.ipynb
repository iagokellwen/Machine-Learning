PCA com Scikit-Learn
o	Utilize o PCA da biblioteca Scikit-Learn em um conjunto de dados como o Iris. Compare a variância explicada pelos componentes principais e visualize os dados reduzidos.


```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
from sklearn.datasets import load_iris
from sklearn.decomposition import PCA
```

```python
# 1. Carregar o conjunto de dados Iris
iris = load_iris()
data_iris = iris.data
labels = iris.target
label_names = iris.target_names
```

```python
# 2. Normalizar os dados
scaler = StandardScaler()
data_iris_scaled = scaler.fit_transform(data_iris)
```

```python
# 1. Aplicar o PCA com 2 componentes principais
pca = PCA(n_components=2)
data_iris_pca_sklearn = pca.fit_transform(data_iris_scaled)
```

```python
# 3. Aplicar o PCA com 2 componentes principais
pca = PCA(n_components=2)
data_iris_pca_sklearn = pca.fit_transform(data_iris_scaled)
```

```python
# 4. Variância explicada por cada componente principal
explained_variance = pca.explained_variance_ratio_
```

```python
# 5. Visualizar os dados reduzidos em 2D
plt.figure(figsize=(8, 6))
for i, label in enumerate(np.unique(labels)):
    subset = data_iris_pca_sklearn[labels == label]
    plt.scatter(subset[:, 0], subset[:, 1], label=label_names[label], alpha=0.7)

plt.title('Dados Iris Após PCA (2D) com Scikit-Learn')
plt.xlabel(f'Componente Principal 1 ({explained_variance[0]:.2f} de variação explicada)')
plt.ylabel(f'Componente Principal 2 ({explained_variance[1]:.2f} de variação explicada)')
plt.legend(loc='best')
plt.tight_layout()
plt.show()

# Mostrar a variância explicada pelos componentes
print(f"Variância explicada pelo Componente 1: {explained_variance[0]:.2f}")
print(f"Variância explicada pelo Componente 2: {explained_variance[1]:.2f}")
```
