LDA (Análise Discriminante Linear)
o	Implemente o LDA para um conjunto de dados que possua mais de duas classes. Visualize a separação entre as classes após a transformação dos dados.

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis as LDA
from sklearn.preprocessing import StandardScaler
```

```python
# 1. Carregar o conjunto de dados Iris
iris = load_iris()
data_iris = iris.data
labels = iris.target
label_names = iris.target_names
```

```python
# 2. Normalizar os dados
scaler = StandardScaler()
data_iris_scaled = scaler.fit_transform(data_iris)
```

```python
# 3. Aplicar o LDA
lda = LDA(n_components=2)  # Queremos reduzir para 2 dimensões
data_iris_lda = lda.fit_transform(data_iris_scaled, labels)
```

```python
# 4. Visualizar os dados projetados em 2D
plt.figure(figsize=(8, 6))
for i, label in enumerate(np.unique(labels)):
    plt.scatter(data_iris_lda[labels == label, 0], data_iris_lda[labels == label, 1],
                label=label_names[label], alpha=0.7)

plt.title('Dados Iris Após LDA (2D)')
plt.xlabel('Discriminante Linear 1')
plt.ylabel('Discriminante Linear 2')
plt.legend(loc='best')
plt.tight_layout()
plt.show()
```
