Análise de Resultados de PCA
o	Após aplicar PCA em um conjunto de dados, analise a contribuição de cada componente principal e como isso afeta a interpretação dos dados.


```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
```

```python
# 1. Carregar o conjunto de dados Iris
iris = load_iris()
X = iris.data
y = iris.target
target_names = iris.target_names
```

```python
# 2. Normalizar os dados
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
display(X_scaled)
```

```python
# 3. Aplicar PCA
pca = PCA()
X_pca = pca.fit_transform(X_scaled)
```

```python
# 4. Analisar a variância explicada
explained_variance = pca.explained_variance_ratio_
```

```python
# Exibir a variância explicada por cada componente
print("Variância explicada por cada componente principal:")
for i, variance in enumerate(explained_variance):
    print(f"Componente {i + 1}: {variance * 100:.2f}%")
```

```python
# 5. Visualizar os Componentes Principais (usando os 2 primeiros componentes)
pca_2d = PCA(n_components=2)
X_pca_2d = pca_2d.fit_transform(X_scaled)

plt.figure(figsize=(8, 6))
for i, target_name in enumerate(target_names):
    plt.scatter(X_pca_2d[y == i, 0], X_pca_2d[y == i, 1], label=target_name)
plt.title('PCA dos Dados Iris')
plt.xlabel('Componente Principal 1')
plt.ylabel('Componente Principal 2')
plt.legend()
plt.grid()
plt.show()
```
